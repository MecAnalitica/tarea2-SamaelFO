Tenemos la Ecuacion Diferencial $$\ddot{\theta}+\mu \dot{\theta}+\mu_{0}sin\theta = Acos(\omega_{0}t)$$
La cual es una ecuacion diferencial no lineal inhomogenea, con una fuerza externa oscilante.

Primer caso es el metodo apróximado donde para angulos pequeños el $$sen \theta = \theta$$



```python
from matplotlib import pyplot
import matplotlib.pyplot as plt
import numpy as np
import sympy 
from scipy import integrate

#ecuacion lineal forzada

def amp(w):
    return 1/np.sqrt((1-w**2)**2+((0.05)*w)**2)

w=np.linspace(0,2,1000)
# Valores del eje fe que toma el gráfico.
Fe = range(0, 10)
# Graficar ambas funciones.
plt.plot(w,amp(w))
plt.title("Amplitud como función de la frecuencia de la fuerza externa", fontsize = 15)
plt.ylabel("Amplitud",fontsize = 18)
plt.xlabel("Fuerza Externa", fontsize = 18)
plt.grid()
# Establecer el color de los ejes.
pyplot.axhline(0, color="black")
pyplot.axvline(0, color="black")
plt.show()
```

b) Grafiquemos la Fase de la solución como función de la frecuencia de la fuerza externa





```python
from matplotlib import pyplot #definimos nuestras librerias a usar
import matplotlib.pyplot as plt
import numpy as np
import sympy 
from scipy import integrate

#ecuacion lineal forzada

def delta(w): #la definición de la frecuencia en función del tiempo esta dado como:
    return np.arctan((0.05*w)/(1-w))

w=np.linspace(0,2,1000)
# Valores del eje fe que toma el gráfico.
Fe = range(0, 10)

# Graficar ambas funciones.
plt.plot(w,delta(w))
plt.title(" Fase de la solución como función de la frecuencia de la fuerza externa", fontsize = 15)
plt.ylabel("Fase",fontsize = 18)
plt.xlabel("Frecuencia Fuerza externa", fontsize = 18)
plt.grid()
# Establecer el color de los ejes.
pyplot.axhline(0, color="blue")
pyplot.axvline(0, color="blue")
plt.show()
```

para la condicion incial A de w=0.70


```python
from matplotlib import pyplot
import matplotlib.pyplot as plt
import numpy as np
import sympy 
from scipy import integrate
import math
t=np.linspace(0,30,1000)
mu = 0.05
w =0.70
delt = np.arctan((mu*w)/(1-w**2))


#tenemos la ecuación para la posicion en funcion del tiempo (theta)
def pos(t,mu,w,delt):
    return np.exp((-mu/2)*t)*(np.cos(np.sqrt(1 - (mu/4)**2)*t)) + (1/(np.sqrt((1-w**2)+ (mu**2)*(w**2)))*np.cos(w*t - delt)) 



#tenemos la ecuación general para la velocidad angular con sus respectivos valores.                                                    
def vang(t,mu,w,delt):                                                
    return -np.exp((-mu/2)*t)*((mu/2)*np.cos(np.sqrt(1-(mu**2/4))*t) + np.sqrt(1-(mu**2/4))*np.sin(np.sqrt(1-(mu**2/4))*t)) + w/(np.sqrt((1-w**2)+ (mu**2)*(w**2)))*np.sin(w*t - delt)   
                                
   




#se definen las graficas en figuras
plt.figure()


# Valores del eje que toma el gráfico, color y nombre en los respectivos ejes.
plt.plot(t,pos(t,mu,w,delt))
plt.title(" Posición del pendulo (theta) como función de t", fontsize = 13)
plt.ylabel("posición",fontsize = 15)
plt.xlabel("Tiempo", fontsize = 15)
plt.grid()
# Establecer el color de los ejes.
pyplot.axhline(0, color="black")
pyplot.axvline(0, color="black")
plt.show()

#segunda grafica
plt.plot(t,vang(t,mu,w,delt),'red')
plt.title(" Velocidad Angular como función de t", fontsize = 15)
plt.ylabel("Velocidad Angular",fontsize = 15)
plt.xlabel("Tiempo", fontsize = 15)
plt.grid()
# Establecer el color de los ejes.
pyplot.axhline(0, color="blue")
pyplot.axvline(0, color="blue")
plt.show()
```
